```python
#6. Write a function that checks if a number is even or odd.
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(9))
```

```python
# 7. Write a function that returns the length of a string.
def string_length(s):
    return len(s)

print(string_length("Hello"))
```

```python
# 8. Write a function that converts Celsius to Fahrenheit.
def celsius_to_fahrenheit(c):
    return (c * 1.8) + 32

print(celsius_to_fahrenheit(25))
```

```python
#9. Write a function that takes a list of numbers and returns their average.
def average_list(numbers):
    return sum(numbers) / len(numbers)

print(average_list([2, 4, 6, 8]))
```

```python
#10. Write a function that counts vowels in a string.
def vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

print(vowels("My name is Maryam"))
```

```python
#11. Write a function that takes a list and returns a new list with unique values.
def unique_list(lst):
    return list(set(lst))

print(unique_list([1, 2, 2, 3, 4, 4]))
```

```python
#12. Write a function that checks if a word is a palindrome.
def is_palindrome(word):
    return word == word[::-1]

print(is_palindrome("level"))
```

```python
#13. Write a function that calculates simple interest.
def simple_interest(principal, rate, time):
    return (principal * rate * time) / 100

print(simple_interest(1000, 5, 2))
```

```python
#14. Write a function with a default parameter for tax rate.
def calculate_tax(amount, tax_rate=5):
    return amount * tax_rate / 100

print(calculate_tax(1000))
print(calculate_tax(1000, 10))
```

```python
#15. Write a function that returns the factorial of a number.
def factor(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print(factor(7))
```

```python
#25. Write a function that accepts score and returns grade using default grading scale.
def get_grade(score):
    if score >= 70:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    elif score >= 45:
        return "D"
    else:
        return "F"

print(get_grade(41))
```

```python
#26. Write a function that counts words in a sentence.
def word_count(sentence):
    return len(sentence.split())

print(word_count("Roses are red"))
```

```python
#27. Write a function that checks if a number is prime.
def prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(prime(7))
```

```python
#28. Write a function that returns the sum of digits of a number.
def sum_digits(n):
    return sum(int(digit) for digit in str(abs(n)))

print(sum_digits(5678))
```

```python
#29. Write a function that accepts a list of prices and applies a default discount.
def default_discount(prices, discount=20):
    return [price * (1 - discount/100) for price in prices]

print(default_discount([100, 200, 300]))
```

```python
#30. Write a function that formats and returns a full name from first and last name.
def full_name(first, last):
    return f"{first} {last}"

print(full_name("fatima", "Mustapha"))
```
